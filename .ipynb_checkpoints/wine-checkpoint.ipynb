### Logistic Regression with Scaling

```python
good_wine = lambda x: 1 if x >= 7 else 0
```

```python
wine_data["Quality"]= wine_data["quality"].apply(good_wine)
wine_data
```

```python
new_wine_data= wine_data.drop(["quality"], axis= 1)
```

```python
new_wine_data.sample(10)
```

```python
def Log_Reg(X, Y, train_size):

    # Division for Training and Testing 
    from sklearn.model_selection import train_test_split
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = train_size, random_state = 21, stratify = Y)

    # Scaling    
    from sklearn.preprocessing import MinMaxScaler
    sc_obj = StandardScaler()
    X_train_sc = sc_obj.fit_transform(X_train)
    X_test_sc = sc_obj.transform(X_test)

    # Model Loading
    from sklearn.linear_model import LogisticRegression
    log_reg = LogisticRegression()

    # Model Fitting/Training
    log_reg.fit(X_train_sc, Y_train)

    m = log_reg.coef_
    c = log_reg.intercept_

    # Prediction
    Y_pred = log_reg.predict(X_test_sc)

    # Evaluation
    from sklearn import metrics
    cls_rpt = metrics.classification_report(Y_test, Y_pred)
    print(f"Classification Report:\n{cls_rpt}")
    
    conf_mat = metrics.confusion_matrix(Y_test, Y_pred)   # Confusion Matrix
    print(f"\nConfusion Matrix:\n{conf_mat}")

    import matplotlib.pyplot as plt
    disp = metrics.ConfusionMatrixDisplay(conf_mat)   # Confusion Matrix Visualisation   # TN FP
    disp.plot()                                                                          # FN TP
    plt.show()

    # F-Beta Score
    F1 = metrics.fbeta_score(Y_test, Y_pred, beta = 1)
    print(f"\nF-1 Score: {F1}")
    
    fpr, tpr, threshold = metrics.roc_curve(Y_test, Y_pred)
    
    # Calculation of TPR and FPR for various thresholds
    import numpy as np
    TPR_list, FPR_list = [], []
    
    for i in np.arange(0, 1.1, 0.1):
        Y_pred_prob = log_reg.predict_proba(X_test_sc)
        Y_pred_prob_one = Y_pred_prob[ : , 1]
        
        for j in range(0, len(Y_pred_prob_one)):
            
            if Y_pred_prob_one[j] >= i:
                Y_pred_prob_one[j] = 1
            else:
                Y_pred_prob_one[j] = 0
        
        tn, fp, fn, tp = metrics.confusion_matrix(Y_test, Y_pred_prob_one).ravel()

        tpr = tp/(tp + fn)
        TPR_list.append(tpr)
        
        fpr = fp/(tn + fp)
        FPR_list.append(fpr)
   
        
    import pandas as pd
    # Dataframe of TPR and FPR values changing w.r.t thresholds
    df = pd.DataFrame({"TPR": TPR_list, "FPR": FPR_list, "Threshold": np.arange(0, 1.1, 0.1)})
    print(f"\nDifferent values of TPR and FPR varying w.r.t Threshold values:")
    display(df)


    # ROC(Receiver Operating Characteristics Curve)
    plt.plot(df["FPR"], df["TPR"], "co--", linewidth = 1, markersize = 4, label = "ROC")
    plt.plot([0, 1], [0, 1], "k--", label = "Base Line")
    plt.xlabel("False Positive Rate (1 - Specificity)")
    plt.ylabel("True Positive Rate (Sensitivity)")
    plt.title("ROC Curve")
    plt.legend(loc = "upper left")
    plt.show()

    # AUC(Area Under Curve)
    auc = metrics.roc_auc_score(Y_test, Y_pred)
    print(f"AUC = {auc}")
```

### Logistic Regression Model without Scaling

```python
x= new_wine_data.iloc[: , : -1]
y= new_wine_data.iloc[:, -1]
```

```python
# Division for Training and Testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, train_size = 0.8, random_state = 21, stratify = y)
```

```python
# Model Loading

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()    
```

```python
# Model Fitting/Training

log_reg.fit(x_train, y_train)
```

```python
# Prediction

y_pred = log_reg.predict(x_test)
```

```python
# Evaluation

from sklearn import metrics
cls_rpt = metrics.classification_report(y_test, y_pred)
print(f"Classification Report:\n{cls_rpt}")
    
conf_mat = metrics.confusion_matrix(y_test, y_pred)   # Confusion Matrix
print(f"\nConfusion Matrix:\n{conf_mat}")

import matplotlib.pyplot as plt
disp = metrics.ConfusionMatrixDisplay(conf_mat)   # Confusion Matrix Visualisation   # TN FP
disp.plot()                                                                          # FN TP
plt.show()

# F-Beta Score
F1 = metrics.fbeta_score(y_test, y_pred, beta = 1)
print(f"\nF-1 Score: {F1}")
```
